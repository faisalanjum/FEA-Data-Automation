### Can see predictibility of FEA columns seperately first and then combine with other variables?

```python
obj=DataRetreivalController()

fea_df = pd.DataFrame()

for mapper in model_mapper_list:
    try:
        data=obj.query_data(mapper=mapper)
        df=pd.DataFrame(data)
        df = df.pivot_table(index=['date'],columns=['description'],values=['value']).sort_index(ascending=False)
        fea_df = pd.concat([fea_df, df], axis=1).fillna(method='ffill')
    except:
        print(mapper)
        pass
fea_df = fea_df.fillna(method='bfill')
fea_df
```
